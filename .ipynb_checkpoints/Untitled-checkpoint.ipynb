# shift function moves the  1 cell below of the value 

```python
adani.shift(periods=1)
```

```python
adani["lag1"]=adani.shift(periods=1)
```

### new column has been added

```python
adani
```

#### subtracting the original vs one row below

```python
adani["ADANIENT.NS"].sub(adani.lag1)
```

```python
adani["Diff"]=adani["ADANIENT.NS"].sub(adani.lag1)
```

```python
adani
```

### dividing the closing data(ADANIENT.NS) with lag1 data and then again subtracting it with 1 and multiplying by 100

```python
adani["%change"]=adani["ADANIENT.NS"].div(adani.lag1).sub(1).mul(100)
```

```python
adani
```

### just using diff function instead of sub which directly subtract no need to make any other column

```python
adani["diif2"] = adani["ADANIENT.NS"].diff(periods=1)
```

```python
adani
```

#### using pct_change instead of percentage change means no need to take diif and subtract one instead use direct function

```python
adani["%change2"] =adani["ADANIENT.NS"].pct_change(periods=1).mul(100)
```

```python
adani
```

### using delete function to delete the whole column

```python
del adani["lag1"]
```

```python
adani
```

#### using drop function instead to delete everything

```python
adani.drop(["Diff", "%change","diif2"], axis=1, inplace=True)
```

```python
adani
```

### using rename function

```python
adani.rename(columns={'%change2':'change'}, inplace= True)
```

```python
adani
```

```python
ret=close.pct_change()
```

```python
ret.describe().T
```

```python
summary=ret.describe().T.loc[:,("mean","std")]
summary
```

```python
summary["mean"]=summary["mean"]*252
summary["std"]=summary["std"]*np.sqrt(252)
summary
```

#####  for i in summary.index means start from i th index 
##### like it starts from acc to tatamotors 

##### plt.annotate(i,xy=(summary.loc[i,"std"]+0.002,summary.loc[i,"mean"]+0.002),size=15)
##### means plot xy as x and y as with i index and continue in for loop for next row

```python
fruits=["aaple","banana","mango"]
for i in fruits:
    print(i)
```

```python
fruits[1]
```

```python
summary.plot.scatter(x="std",y="mean",figsize=(12,8),s=50,fontsize=15)
for i in summary.index:
    plt.annotate(i,xy=(summary.loc[i,"std"]+0.002,summary.loc[i,"mean"]+0.002),size=15)
plt.xlabel("annual risk(std)",fontsize=15)
plt.ylabel("annual returns",fontsize=15)
plt.title("Risk/Return",fontsize=25)
```
